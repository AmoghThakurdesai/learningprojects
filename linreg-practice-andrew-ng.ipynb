```python
houseprices = pd.read_csv("https://raw.githubusercontent.com/ywchiu/riii/master/data/house-prices.csv",sep=",")
houseprices = houseprices[["SqFt","Price"]]
houseprices = houseprices.to_numpy()
houseprices = houseprices.T
print(houseprices.shape)
houseprices
```

```python
m = houseprices.shape[1]
X = np.vstack((houseprices[0],np.ones((m)))).T
Y = houseprices[1].reshape(m,1)
theta = np.random.random((2,1))
X[:,0],Y.shape
```

```python
def model(t,x):
    return x.dot(t)
def error(t,x,y):
    print("t",t,"x",x,"y",y)
    return model(t,x) - y 
def cost(t,x,y):
    print("error",error(t,x,y))
    return np.sum(np.square(error(t,x,y)))/(2*m)
def grad_descent(t,x,y,epochs,alpha):
    for i in range(0,epochs):
        temp = np.zeros((2,1))
        temp[0] = t[0]-(alpha*(np.sum(np.multiply(error(t,x,y),x[0,:]))))/m
        temp[1] = t[1]-(alpha*(np.sum(error(t,x,y))))/m
        print("Temp 1",temp[0],"Temp 2",temp[1])
        t = np.array([temp[0],temp[1]]).reshape((2,1))
        print("t",t);
        print("Cost" ,cost(t,x,y))
    return t
```

```python
theta = grad_descent(theta,X,Y,epochs=10,alpha = 0.01)
```
